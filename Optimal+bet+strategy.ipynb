```python
mean_return = .1
volatility =  .2
risk_free = .05

growth_rate_wealth = [] 

for fraction_of_b_star in range(1,21):
    
    b_star =  (mean_return -  risk_free)/(volatility**2)

    fraction_of_b_star =  fraction_of_b_star/10

    b_star = fraction_of_b_star * b_star

    optimal_growth_rate = b_star*mean_return - .5*(b_star**2)*(volatility**2)+(1-b_star)*risk_free
    
    growth_rate_wealth.append(optimal_growth_rate)
    
    #print(fraction_of_b_star,optimal_growth_rate)


%matplotlib inline    

import matplotlib.pyplot as plt
plt.plot(growth_rate_wealth)
plt.ylabel('Growth rate of wealth')
plt.xlabel('Fraction of bstar bet')
```
